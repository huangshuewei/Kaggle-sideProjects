```python
# Read data
train_data = pd.read_csv("/kaggle/input/titanic/train.csv")
test_data = pd.read_csv("/kaggle/input/titanic/test.csv")
train_data.head()
test_data['Sex'].head()

# Replace 'Sex' male and female to 0 and 1
train_data['Sex'] = train_data['Sex'].replace(['male'], 0)
train_data['Sex'] = train_data['Sex'].replace(['female'], 1)
test_data['Sex'] = test_data['Sex'].replace(['male'], 0)
test_data['Sex'] = test_data['Sex'].replace(['female'], 1)
```

```python
# X and y for training
features = ["Pclass", "Sex", "SibSp", "Parch"]
X = pd.get_dummies(train_data[features])
X = X.iloc[:].values
Xtest = pd.get_dummies(test_data[features])
Xtest = Xtest.iloc[:].values
y = train_data['Survived']
y = y.iloc[:].values

# Show data shape
print(X.shape)
print(y.shape)
print(X)
```

```python
from sklearn.ensemble import RandomForestClassifier

# Random Forest Classifier with 100 estimators and 5 maximum depth.
model = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=100)
# Feed data to this model.
model.fit(X, y)

print("Finish training.")
```

```python
# Do prediction.
predictions = model.predict(Xtest)

print(predictions)

output = pd.DataFrame({'PassengerId': test_data.PassengerId, 'Survived': predictions})
output.to_csv('submission.csv', index=False)
print("Your submission was successfully saved!")
```
